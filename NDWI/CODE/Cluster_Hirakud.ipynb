```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
import sklearn.cluster as cluster
import sklearn.metrics as metrics
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
scaler = MinMaxScaler()
scale = scaler.fit_transform(df[['Green','Red','Nir','Swir1','Swir2','NDVI_nir','NDVI_swir1','NDVI_swir2','NDVI_nir','Turbidity','Water_ratio_index']])
df_scale = pd.DataFrame(scale, columns = ['Green','Red','Nir','Swir1','Swir2','NDVI_nir','NDVI_swir1','NDVI_swir2','NDVI_nir','Turbidity','Water_ratio_index']);
df_scale.head(5)
```

```python
km=KMeans(n_clusters=2)
y_predicted = km.fit_predict(df_scale)
y_predicted
```

```python
df_Res = pd.DataFrame(y_predicted)
df_Res[0]
```

```python
df['Cluster'] = df_Res[0]
```
